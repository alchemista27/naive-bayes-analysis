```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google_play_scraper import Sort, reviews, app
import re


# initialize data crawling
reviews, token = reviews(
    app_id='com.nordvpn.android',
    lang='id',
    country='id',
    count=500,
    sort=Sort.MOST_RELEVANT,
    filter_score_with=None
)


print(reviews)
```

```python
# Creating dataframe
raw_data = pd.DataFrame(reviews)
raw_data


# Select data into new dataframe and droping duplicated data
df = raw_data[['userName', 'content', 'score', 'at']]
df = df.drop_duplicates(subset=['content'])
df.duplicated().sum()


# Checking missing value and dropping missing value
df.isna().sum()


# Save the initial data
df.to_csv('nord-vpn-reviews.csv', index=False)
```
