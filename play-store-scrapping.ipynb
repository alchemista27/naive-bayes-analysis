```python
# Import Libraries
import pandas as pd
import numpy as np
from google_play_scraper import app, Sort, reviews
```

```python
# Initiating data crawling
rvws, token = reviews(
    app_id='free.vpn.unblock.proxy.turbovpn',
    lang='en',
    country='us',
    count=10000,
    sort= Sort.NEWEST,
    filter_score_with= None
)


# Converting rvws into DataFrame
revs = pd.DataFrame(rvws)
revs.head()
```

```python
# Slicing the data into used data
df = revs[['userName', 'content', 'score', 'at']]
df.head()


# Checking the dtype
df.info()


# Save the data into csv
df.to_csv('reviews-turbo-vpn.csv', index=False)
```
