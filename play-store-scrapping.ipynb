```python
# Import Libraries
import pandas as pd
import numpy as np
from google_play_scraper import app, Sort, reviews
```

```python
# Initiating data crawling
langs = ['en', 'id']
countries = ['id', 'uk', 'us']
data = {}

for lang in langs:
    for country in countries:
        review, token = reviews(
            app_id='free.vpn.unblock.proxy.turbovpn',
            lang=lang,
            country=country,
            count=10000,
            sort= Sort.MOST_RELEVANT,
            filter_score_with= None
        )
        data[lang] = review
print(data['id'])
print(data['en'])
```

```python
# Creating dataframe
data_en = pd.DataFrame(data['en'])
data_id = pd.DataFrame(data['id'])


# Concatenate the data
data = pd.concat([data_id, data_en])
data.head()
```

```python
# Creating df containing reviewId, userName, content, score, and at
df = data[['reviewId', 'userName', 'content', 'score', 'at']]
df.head()


# 
```
